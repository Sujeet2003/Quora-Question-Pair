```python
# Basic text preprocessing

from bs4 import BeautifulSoup
import re
from nltk.stem import SnowballStemmer

def preprocess(question):

  # lower casing
  q = str(question).lower().strip()

  # html tags
  q = BeautifulSoup(q).get_text()

  # replacing the certain characters with it's corresponding strings
  q = q.replace("%", "percent ")
  q = q.replace("$", "dollar ")
  q = q.replace("₹", "rupee ")
  q = q.replace("€", "euro ")
  q = q.replace("@", "at ")

  # The pattern `[math]` appearns in the datasets, so replacing to blank as: ""
  q = q.replace("[math]", "")

  # replacing numbers with it's corresponding string values
  q = q.replace(",000,000,000", "b ")
  q = q.replace(",000,000", "m ")
  q = q.replace(",000", "k ")
  q = re.sub(r"([0-9]+)000000000", r"\1b", q)
  q = re.sub(r"([0-9]+)000000", r"\1m", q)
  q = re.sub(r"([0-9]+)000", r"\1k", q)

  # decontraction of words
  q_decontracted = []
  for word in q.split():
    if word in contractions:
      word = contractions[word]

    q_decontracted.append(word)
  q = " ".join(q_decontracted)

  # replacing some decontracted words
  q = q.replace("'ve", "have")
  q = q.replace("n't", "not")
  q = q.replace("'re", "are")
  q = q.replace("'ll'", "will")

  # removing punctations
  pattern = re.compile("\W")
  q = re.sub(pattern, " ", q).strip()

  # stemming or lemmatizer
  snowball = SnowballStemmer("english")
  words = q.split()
  stemmed_word_list = [snowball.stem(word) for word in words]
  q = " ".join(stemmed_word_list)

  return q
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/Datasets/quora.csv"

batch_size = 10000

def basic_preprocessing(chunk):

    chunk['question1'] = chunk['question1'].apply(preprocess)
    chunk['question2'] = chunk['question2'].apply(preprocess)

    return chunk

with pd.read_csv(file_path, chunksize=batch_size) as reader:
    for i, chunk in enumerate(reader):
        print(f"Processing chunk {i + 1}...")

        # Preprocess the current chunk
        processed_chunk = basic_preprocessing(chunk)

        # Write to CSV (append to avoid overwriting)
        if i == 0:
            # Write header only for the first chunk
            processed_chunk.to_csv(output_file_path, index=False, mode='w')
        else:
            # Append without writing the header
            processed_chunk.to_csv(output_file_path, index=False, mode='a', header=False)

print("Processing complete! Processed data saved to:", output_file_path)
```

### Extra/Intermediate Feature Engineering

### Adding `7 columns` with extra features as `question length`, `words count`, `total words count`, `common words`, `shared words`

```python
def extra_preprocessing(row):

  extra_features = []

  q1 = str(row['question1'])
  q2 = str(row['question2'])

  # length of question1 & question2
  extra_features.append(len(str(q1)))
  extra_features.append(len(str(q2)))

  # words count of question1 & question2
  q1_word_count = len(q1.split(" "))
  q2_word_count = len(q2.split(" "))
  extra_features.append(q1_word_count)
  extra_features.append(q2_word_count)

  # total words
  total_words_count = q1_word_count + q2_word_count
  extra_features.append(total_words_count)

  # common words
  q1_words = set(map(lambda x: x.strip(), q1.split()))
  q2_words = set(map(lambda x: x.strip(), q2.split()))
  common_words_count = len(q1_words & q2_words)
  extra_features.append(common_words_count)

  # shared words
  shared_words = np.round((common_words_count/total_words_count), 2)
  extra_features.append(shared_words)

  return extra_features
```

```python
output_file_path = "extra_preprocessed_data.csv"
```

```python
with pd.read_csv(file_path, chunksize=10000) as reader:
    for i, chunk in enumerate(reader):
        print(f"Processing chunk {i + 1}...")

        # Apply the `extra_preprocessing` function to each row of the chunk
        extra_features_df = chunk.apply(extra_preprocessing, axis=1, result_type='expand')

        # Rename the new columns for clarity
        extra_features_df.columns = [
            'q1_len', 'q2_len', 'q1_word_count', 'q2_word_count',
            'total_word_count', 'common_words_count', 'shared_words'
        ]

        # Concatenate the new features to the original chunk
        chunk = pd.concat([chunk, extra_features_df], axis=1)

        # Write the processed chunk to a CSV file
        if i == 0:
            # For the first chunk, write the header
            chunk.to_csv(output_file_path, index=False, mode='w')
        else:
            # For subsequent chunks, append without the header
            chunk.to_csv(output_file_path, index=False, mode='a', header=False)

print("Processing complete! Processed data saved to:", output_file_path)
```
